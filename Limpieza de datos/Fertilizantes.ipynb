```python
# importaciones
import pandas as pd
df = pd.read_csv('../Datos/Actual/Fertilizantes.csv', sep=',')
df.head()
```

```python
# limpieza de duplicados
print(len(df.index))
df_fertilizante = df.drop_duplicates(subset=['Nombre del Producto', 'Tipo de Cultivo'])
row_writter  = len(df_fertilizante.index)
```

```python
path_procces = '../Datos/Procesados/Fertilizantes.csv'
df_fertilizante.to_csv(path_procces ,sep=';', index=False)
print(f'Se guardaron {row_writter} registros en la ruta {path_procces}')
```
